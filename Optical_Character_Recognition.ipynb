#Importing the needed libraries for the task

```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize'] = 8, 16
```

```python
reader = easyocr.Reader(['en'])
```

* The Reader class is the base class for EasyOCR which contains a list of language codes and other parameters such as GPU that is by default set to True.

#Reading image


```python
Image("picture.jpg")
```

* This is the cover page of Nora Barrett's novel "THE KINGS OF THE DRUGS"

```python
output = reader.readtext('picture.jpg')
output
```

* The output shows 4 bounding box coordinates(x,y) of the text along with the identified text and confidence score.

# Selecting the text " THE KINGS "

```python
cord = output[-2][0]
```

```python
x_min, y_min = [int(min(idx)) for idx in zip(*cord)]
```

```python
x_max, y_max = [int(max(idx)) for idx in zip(*cord)]
```

```python
image = cv2.imread('picture.jpg')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
